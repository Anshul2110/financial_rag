```python
import json
import os
import pandas as pd
import numpy as np
```

```python
## reading the processed json files for 10-k files to create a sigle json file
processed_data_dir = "../data/processed"
all_data = []
for filename in os.listdir(processed_data_dir):
    if filename.endswith(".json"):
        with open(os.path.join(processed_data_dir, filename), "r") as f:
            data = json.load(f)
            all_data.append(data)
```

```python
for data in all_data:
    data['file_name'] = data['file_name'].split('-')[0]
```

```python
all_data
```

```python
# Save the combined data to a single JSON file
with open("../data/combined_10k_data.json", "w") as f:
    json.dump(all_data, f)
```

```python
## counting average number of words in risk section and mda section
for data in all_data:
    risk_words = len(data['risk_factors'].split())
    mda_words = len(data['management_discussion_and_analysis'].split())
    print(f"File: {data['file_name']}, Risk Words: {risk_words}, MDA Words: {mda_words}")
```
