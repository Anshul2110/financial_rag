```python
import os
import re
from bs4 import BeautifulSoup
from pathlib import Path
import json
```

```python
raw_data_dir = Path("../data/raw")
processed_data_dir = Path("../data/processed")  
processed_data_dir.mkdir(parents=True, exist_ok=True)
```

```python
def cleanText(text):
    text = re.sub(r'\n+', '\n', text)  # Replace multiple newlines with single newline
    text = re.sub(r'[ \t]+', ' ', text)  # Replace multiple spaces/tabs with single space
    return text.strip()
```

```python
def extract_section(text,start_pattern,end_pattern):
    matches = re.findall(f'{start_pattern}(.*?){end_pattern}', text, flags= re.IGNORECASE | re.DOTALL)
    if not matches:
        return ""
    return max(matches, key=len)
```

```python
def process_file(file_path):
    with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
        html = f.read()

    soup = BeautifulSoup(html, "lxml")
    text = soup.get_text(separator="\n")
    text = cleanText(text)

    risk_text = extract_section(
        text,
        r"Item\s+1A\.?\s*Risk\s+Factors",
        r"Item\s+1B"
    )

    mda_text = extract_section(
        text,
        r"Item\s+7\.?\s*Management",
        r"Item\s+7A"
    )

    return risk_text, mda_text

```

```python
for file in raw_data_dir.glob("*.html"):
    print(f"Processing {file.name}")
    risk, mda = process_file(file)

    output = {
        "file_name": file.name,
        "risk_factors": risk,
        "management_discussion_and_analysis": mda
    }

    output_path = processed_data_dir / f"{file.stem}.json"

    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(output, f, indent=2)

    print(f"Saved {output_path}")
```
